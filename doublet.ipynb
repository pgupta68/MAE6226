```python
import math
import numpy
from matplotlib import pyplot
%matplotlib inline
```

```python
N = 50                                
x_start, x_end = -2.0, 2.0            
y_start, y_end = -1.0, 1.0            
x = numpy.linspace(x_start, x_end, N)    
y = numpy.linspace(y_start, y_end, N)    
X, Y = numpy.meshgrid(x, y)              
k = 1.0                        
x_d, y_d = 0.0, 0.0
width = 10
height = (y_end - y_start) / (x_end - x_start) * width
```

```python
u = (- k / (2 * math.pi) *((X)**2 - (Y)**2) /((X)**2 + (Y)**2)**2)
v = (- k / (2 * math.pi) * 2 * (X) * (Y) / ((X)**2 + (Y)**2)**2)
psi = - k / (2 * math.pi) * (Y) / ((X)**2 + (Y)**2)
```

```python
pyplot.figure(figsize=(width, height))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.streamplot(X, Y, u, v, density=3, linewidth=1, arrowsize=1, arrowstyle='->')
pyplot.scatter(x_d, y_d, color='green', s=80, marker='o');
```

```python
u_infinite=1.0
u_free = u_infinite * numpy.ones((N, N), dtype=float)
v_free = numpy.zeros((N, N), dtype=float)
psi_free = u_infinite * Y
```

```python
u = u_free + u
v = v_free + v
psi_total = psi_free + psi

pyplot.figure(figsize=(width, height))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.streamplot(X, Y, u, v,
                  density=2, linewidth=1, arrowsize=1, arrowstyle='->')
pyplot.contour(X, Y, psi_total,
               levels=[0.], colors='red', linewidths=2, linestyles='solid')
pyplot.scatter(x_d, y_d, color='black', s=80, marker='o')

x_stag1 = +math.sqrt(k / (2 * math.pi * u_infinite))
y_stag1 = 0.0
x_stag2 = -math.sqrt(k / (2 * math.pi * u_infinite)) 
y_stag2 = 0.0
pyplot.scatter([x_stag1, x_stag2], [y_stag1, y_stag2], color='g', s=80, marker='o');
```

```python
cp = 1.0 - (u**2 + v**2) / u_infinite**2
pyplot.figure(figsize=(1.1 * width, height))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
contf = pyplot.contourf(X, Y, cp,
                        levels=numpy.linspace(-2.0, 1.0, 100), extend='both')
cbar = pyplot.colorbar(contf)
cbar.set_label('yellow', fontsize=16)
cbar.set_ticks([-2.0, -1.0, 0.0, 1.0])
pyplot.scatter(x_d, y_d,
               color='red', s=80, marker='o')
pyplot.contour(X,Y,psi,
               levels=[0.], colors='black', linewidths=2, linestyles='solid')
pyplot.scatter([x_stag1, x_stag2], [y_stag1, y_stag2],
               color='b', s=80, marker='o');
```
